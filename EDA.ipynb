# **Step 1: Loading the Dataset, Exploring Dimensions, Summary, and Data Handling**

## 1.1 Load Dataset

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bcancer.csv'
df = pd.read_csv(url)

# View the first few rows of the dataset
df.head()
```

## 1.2 Explore Dimensions and Summary

```python
# Check dataset dimensions
print(f"The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# Basic information of the dataset
df.info()

# Statistical summary of numerical features
df.describe()
```

## 1.3 Data Handling: Check and fill missing values in numeric columns with median and non-numeric columns with mode.


```python
# Checking for missing values
print(df.isnull().sum())

# Separating numeric columns from non-numeric ones
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
non_numeric_cols = df.select_dtypes(exclude=['float64', 'int64']).columns

# Fill missing values for numeric columns with the median
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Fill missing values for non-numeric columns (if needed) with the mode (most frequent value)
for col in non_numeric_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Check if all missing values are filled
print(df.isnull().sum())
```

```python
# Drop the 'Unnamed: 32' column
df.drop('Unnamed: 32', axis=1, inplace=True)
```

# **Step 2: Data Cleaning and Handling**
## 2.1 Removing Duplicates

```python
# Check for duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Remove duplicates
df = df.drop_duplicates()
print(f"Shape after removing duplicates: {df.shape}")
```

## 2.2 Outlier Detection and Filtering

```python
# Detect outliers using boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Boxplot for Outlier Detection (Reg. No.: 21BDS0113)')
plt.xticks(rotation=90)
plt.show()

# Filter out outliers based on certain criteria (example: filtering radius_mean > threshold)
df_filtered = df[df['radius_mean'] < 30]
print(f"Shape after filtering out outliers: {df_filtered.shape}")
```

# **Step 3: Univariate Analysis**
## 3.1 Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style and palette
sns.set(style="whitegrid")
plt.style.use('ggplot')

# Define a purple color for the histograms
color = '#6a0dad'  # Purple color

# Select only numeric columns
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Calculate number of rows needed based on the number of features
num_features = len(df_numeric.columns)
num_cols = 3  # Number of columns in the subplot grid
num_rows = (num_features // num_cols) + (num_features % num_cols > 0)

# Create subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 5 * num_rows))
axes = axes.flatten()  # Flatten the axes array for easy iteration

# Plot each feature's histogram
for i, column in enumerate(df_numeric.columns):
    ax = axes[i]
    df_numeric[column].plot(kind='hist', ax=ax, bins=20, color=color, edgecolor='black', linewidth=1.0)
    ax.set_title(column, fontsize=12, weight='bold')
    ax.set_xlabel('Values', fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)

# Remove any unused subplots (if the grid is larger than the number of plots)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('Histogram of All Features (Reg. No.: 21BDS0113)', fontsize=22, weight='bold')
plt.subplots_adjust(top=0.95, hspace=0.4, wspace=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style and palette
sns.set(style="whitegrid")
plt.style.use('ggplot')

# Select only numeric columns
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Define a color palette with diverse colors for clarity
palette = sns.color_palette("coolwarm", n_colors=len(df_numeric.columns))

# Create the boxplot
plt.figure(figsize=(15, 10))
sns.boxplot(data=df_numeric, palette=palette, width=0.5)

# Enhance plot aesthetics
plt.title('Boxplot for All Features (Reg. No.: 21BDS0113)', fontsize=22, weight='bold')
plt.xlabel('Features', fontsize=15)
plt.ylabel('Values', fontsize=15)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # grid lines for better readability

# Show plot
plt.tight_layout()  # Adjust layout to avoid clipping of labels
plt.show()
```

# **Step 4: Bivariate Analysis**
## 4.1 Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# Define a custom purple-pink colormap with more distinct gradient
colors = ["#f7e1ea", "#e08cb5", "#b96588", "#6D28D9"]  # Lighter pink to dark purple for contrast
cmap = LinearSegmentedColormap.from_list("purple_pink", colors, N=256)

# Select the relevant columns for analysis
columns_of_interest = [
    'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
    'smoothness_mean', 'compactness_mean', 'concavity_mean',
    'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean'
]
df_relevant = df[columns_of_interest]

# Compute the correlation matrix for the selected columns
corr = df_relevant.corr()

# Create the heatmap
plt.figure(figsize=(14, 10))  # Adjusted size for better readability
sns.heatmap(
    corr,
    annot=True,  # Show correlation values
    cmap=cmap,  # Custom purple-pink colormap with better contrast
    fmt='.2f',  # Format annotations
    linewidths=0.5,  # Add separation lines between cells
    vmin=-1, vmax=1,  # Set the color range
    center=0,  # Center the colormap at 0 for better clarity of positive/negative correlations
    cbar_kws={"shrink": .8},  # Adjust color bar size
    annot_kws={"size": 12, "weight": 'bold'}  # Annotation font size and weight
)

# Customize the plot
plt.title('Correlation Matrix Heatmap of Selected Features (Reg. No.: 21BDS0113)', fontsize=22, weight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate x-axis labels for better readability
plt.yticks(rotation=0, fontsize=12)  # Keep y-axis labels horizontal

# Show the plot
plt.tight_layout()  # Adjust layout to fit elements
plt.show()
```

## 7.3 Grouping and Aggregation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (uncomment and set the path if needed)
# df = pd.read_csv('path_to_your_dataset.csv')

# Check for NaN values in the 'diagnosis' column
print("Number of NaN values in 'diagnosis':", df['diagnosis'].isna().sum())

# Fill NaN values or drop rows (choose one)
# Option 1: Fill NaN with a default value (e.g., 0 for Benign)
# df['diagnosis'].fillna('B', inplace=True)  # Assuming 'B' is the default

# Option 2: Drop rows with NaN in 'diagnosis'
df.dropna(subset=['diagnosis'], inplace=True)

# Ensure 'diagnosis' is correctly mapped to numeric (0 for Benign, 1 for Malignant)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0}).astype(int)

# Grouping by 'diagnosis' and calculating the mean for each feature
df_grouped = df.groupby('diagnosis').mean()

# Transpose the grouped DataFrame for better visualization
df_grouped = df_grouped.T

# Set figure size and define custom color palette
plt.figure(figsize=(14, 8))
colors = ["#6A0DAD", "#FF69B4"]  # Dark purple and pink

# Create the bar plot with enhanced aesthetics
bars = plt.bar(df_grouped.index, df_grouped[0], color=colors[0], edgecolor='black', label='Benign')
plt.bar(df_grouped.index, df_grouped[1], bottom=df_grouped[0], color=colors[1], edgecolor='black', label='Malignant')

# Add annotations on top of bars
for i in range(len(df_grouped)):
    plt.text(i, df_grouped[0].iloc[i] + df_grouped[1].iloc[i] - 0.05,
             f'{df_grouped[0].iloc[i]:.2f}', ha='center', fontsize=10, color='black', weight='bold')

# Customize plot appearance
plt.title('Average Values of Features by Diagnosis (21BDS0113)', fontsize=22, weight='bold')
plt.xlabel('Features', fontsize=16)
plt.ylabel('Average Value', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Diagnosis (0: Benign, 1: Malignant)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add horizontal grid lines for better readability

# Display the plot with a neat layout
plt.tight_layout()
plt.show()
```
